```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli, uniform, poisson, norm, lognorm

# Set the style of plots
sns.set(style="whitegrid")

# Parameters for Bernoulli distribution
p = 0.3  # Probability of success

# Generate samples
bernoulli_samples = bernoulli.rvs(p, size=1000)

# Plot histogram and overlay the PMF
plt.figure(figsize=(8, 6))
sns.histplot(bernoulli_samples, bins=2, kde=False, stat="density", color='skyblue', discrete=True)

# Plot the PMF of the Bernoulli distribution
x_vals = [0, 1]
pmf_vals = bernoulli.pmf(x_vals, p)
plt.vlines(x_vals, 0, pmf_vals, colors='red', linewidth=3, label='PMF')
plt.title('Bernoulli Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

The Bernoulli distribution describes a random variable which has only two possible outcomes, typically represented as 0 or 1 for computing purposes such as artificial nerurons.

The gerneral formula is:
P(X=x)=p^x((1-p)^(1−x)), where x∈{0,1}
p is the probability of success or rather the probability that for x, X=1.
(1-p) is the probability of failure,the probability that X=0, for a given x.

Think of it as n=1 or 100%, sum(p)=1 so n=x+(p-x) for a given x.

```python
# Parameters for Uniform distribution
a, b = 0, 10  # Range of the uniform distribution

# Generate samples
uniform_samples = uniform.rvs(loc=a, scale=b-a, size=1000)

# Plot histogram and overlay the PDF
plt.figure(figsize=(8, 6))
sns.histplot(uniform_samples, bins=30, kde=False, stat="density", color='lightgreen')

# Plot the PDF of the Uniform distribution
x_vals = np.linspace(a, b, 1000)
pdf_vals = uniform.pdf(x_vals, loc=a, scale=b-a)
plt.plot(x_vals, pdf_vals, color='red', linewidth=3, label='PDF')
plt.title('Uniform Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

The Uniform distribution is a distribution where all outcomes in a range are equally likely.

This can be done as a weighted adverage of a lot of variables as shown above, or a a super basic "or" statment for deturmining a true or false return, think heads or tails. Though that lends itself towards Bernoulli classification rather than a multivariable uniform distrbution.

The gereral formula for uniform distribution is:
f(x)= (b−a)/1, for x∈[a,b].

Or an very simple adverage calculation across the data range. Where 'a' is the lower bound and 'b' is the upper bound.

```python
# Parameters for Poisson distribution
lambda_ = 5  # Average rate of occurrence

# Generate samples
poisson_samples = poisson.rvs(lambda_, size=1000)

# Plot histogram and overlay the PMF
plt.figure(figsize=(8, 6))
sns.histplot(poisson_samples, bins=20, kde=False, stat="density", color='lightcoral', discrete=True)

# Plot the PMF of the Poisson distribution
x_vals = np.arange(0, np.max(poisson_samples) + 1)
pmf_vals = poisson.pmf(x_vals, lambda_)
plt.vlines(x_vals, 0, pmf_vals, colors='red', linewidth=3, label='PMF')
plt.title('Poisson Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

The Poisson distribution models the number of events occurring in a fixed interval of time or space.

This means that the graph can shift forward or backward across a threshold dependednt on variable input.

The gereral formula for this distribution is:

P(X=x)= (λ*x)*(e^−λ), where x∈{0,1,2,… n}.

λ, calculated seperatly, is the average rate of occurrence of the event.

```python
# Parameters for Gaussian distribution
mu, sigma = 0, 1  # Mean and standard deviation

# Generate samples
gaussian_samples = norm.rvs(loc=mu, scale=sigma, size=1000)

# Plot histogram and overlay the PDF
plt.figure(figsize=(8, 6))
sns.histplot(gaussian_samples, bins=30, kde=False, stat="density", color='lightblue')

# Plot the PDF of the Gaussian distribution
x_vals = np.linspace(-4, 4, 1000)
pdf_vals = norm.pdf(x_vals, loc=mu, scale=sigma)
plt.plot(x_vals, pdf_vals, color='red', linewidth=3, label='PDF')
plt.title('Gaussian (Normal) Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

The Gaussian (or Normal) distribution is a continuous probability distribution defined by the mean (𝜇) and standard deviation (𝜎).

Its PDF (Probability Density Function) is:
f(x)= (1/(σ*Sqrt(2π)))*(e^-(((x-μ)^2)/(2σ^2)))

μ is the mean. Gathered through μ=sum(X(i))/n where X is the array values, n is the array size.

σ is the standard deviation. σ=sqrt((sum(X(i)-μ)^2)/n), where X(i) is each valus of the set, μ is the mean, and n is the array size.


```python
# Parameters for Log-Normal distribution
mu_lognorm, sigma_lognorm = 0, 1  # Mean and standard deviation of the underlying normal distribution

# Generate samples
lognorm_samples = lognorm.rvs(sigma_lognorm, loc=0, scale=np.exp(mu_lognorm), size=1000)

# Plot histogram and overlay the PDF
plt.figure(figsize=(8, 6))
sns.histplot(lognorm_samples, bins=30, kde=False, stat="density", color='lightpink')

# Plot the PDF of the Log-Normal distribution
x_vals = np.linspace(0.01, np.max(lognorm_samples), 1000)
pdf_vals = lognorm.pdf(x_vals, sigma_lognorm, loc=0, scale=np.exp(mu_lognorm))
plt.plot(x_vals, pdf_vals, color='red', linewidth=3, label='PDF')
plt.title('Log-Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

A random variable X follows a log-normal distribution if log(X) is normally distributed.

Its PDF is:

f(x)=1/(xσ*sqrt(2π))*(e^(-((ln(x)−μ)^2)/2σ^2)), for x>0.

μ is the mean of the underlying normal distribution.
σ is the standard deviation of the underlying normal distribution.
*See normal for formulas for mean and standard deviation.

This distribution rises quickly and decays just as rapidly.

```python
# Parameters for White Noise
sample_rate = 1000  # Samples per second
duration = 1  # Duration of the signal in seconds
n_samples = sample_rate * duration  # Total number of samples

# Generate White Noise
white_noise = np.random.normal(0, 1, n_samples)  # Mean 0, Std 1

# Time axis
time = np.linspace(0, duration, n_samples)

# Plot White Noise Signal
plt.figure(figsize=(10, 6))
plt.plot(time, white_noise, color='blue')
plt.title('White Noise')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

The general definition for a white noise signal x(t) is:
x(t)=η(t), where

η(t) is a random variable with a Gaussian distribution (mean 0 and variance σ^2).

The signal has a flat spectral density S(f), meaning the signal's power has a uniform distribution across all frequencies.

This is a measurable random signal often gatherd along with the vital data signal that is meant to be processed and filtered.

```python
# Generate Brownian Noise (Random Walk)
brownian_noise = np.cumsum(np.random.normal(0, 1, n_samples))  # Cumulative sum of white noise

# Plot Brownian Noise Signal
plt.figure(figsize=(10, 6))
plt.plot(time, brownian_noise, color='green')
plt.title('Brownian Noise (Random Walk)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

Brownian noise is generated by taking the cumulative sum of white noise values, making it a random walk. The resulting process has a characteristic power spectral density that is proportional to 1/(f^2), where f is the frequency. This means that lower frequencies have higher energy.

The equation for Brownian motion can be described as:

x(t)=∑η(i), where i∈{1,2,… t}.

ηi is white noise at each time step.
The process is the cumulative sum of the noise values, representing a random walk.




```python
# Function to generate pink noise using Voss-McCartney algorithm
def generate_pink_noise(n_samples):
    # Generate white noise
    white = np.random.normal(0, 1, n_samples)

    # Create a pink noise filter (Voss-McCartney algorithm)
    alpha = 1  # Scaling factor for the 1/f noise
    pink = np.cumsum(white)  # Integrate to generate pink noise

    # Normalize the pink noise
    pink -= np.mean(pink)
    pink /= np.std(pink)

    return pink

# Generate Pink Noise
pink_noise = generate_pink_noise(n_samples)

# Plot Pink Noise Signal
plt.figure(figsize=(10, 6))
plt.plot(time, pink_noise, color='red')
plt.title('Pink Noise')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

Pink noise, also known as 1/f noise, has a power spectral density that is inversely proportional to the frequency. This means that lower frequencies have more power, and the power decreases as 1/f with increasing frequency.

The general definition of pink noise is:
S(f)∝(1/f), where

S(f) is the power spectral density and f is the frequency.

Pink noise can be generated using various methods, such as filtering white noise with a filter that adjusts the frequency response to match the 1/f distribution.
