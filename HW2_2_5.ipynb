```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function J(θ) = (2θ - 3)^2
def J(theta):
    return (2 * theta - 3)**2

# Generate θ values
theta_values = np.linspace(-5, 5, 400)
J_values = J(theta_values)

# Plot J(θ) vs θ
plt.plot(theta_values, J_values, label=r'$J(\theta) = (2\theta - 3)^2$')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$J(\theta)$')
plt.title(r'Plot of $J(\theta) = (2\theta - 3)^2$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the function J(θ) = (2θ - 3)^2
def J(theta):
    return (2 * theta - 3) ** 2

# Derivative of J(θ)
def grad_J(theta):
    return 4 * (2 * theta - 3)

# Gradient Descent function
def gradient_descent(initial_theta, learning_rate, max_iterations, tolerance):
    theta = initial_theta
    for i in range(max_iterations):
        gradient = grad_J(theta)
        theta_new = theta - learning_rate * gradient

        # Check if the change is smaller than tolerance
        if abs(theta_new - theta) < tolerance:
            print(f"Converged after {i+1} iterations.")
            break

        theta = theta_new

    return theta

# Set parameters for gradient descent
initial_theta = 5  # Start with θ = 5
learning_rate = 0.1
max_iterations = 1000
tolerance = 1e-6

# Run gradient descent to find the minimum
theta_min = gradient_descent(initial_theta, learning_rate, max_iterations, tolerance)

# Output the result
print(f"The value of θ that minimizes J(θ) is: {theta_min}")
print(f"The minimum value of J(θ) is: {J(theta_min)}")
```
